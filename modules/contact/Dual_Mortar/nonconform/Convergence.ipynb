```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def inv_list(h):
    invh=[]
    for i in h:
        invh.append(1/i)
    return invh
```

```python
# Finer master

# Non-conforming, 1st order lagrange

dual_h1=[3.849491e-01, 9.623726e-02, 4.811863e-02, 2.405932e-02, 1.202966e-02, 6.014829e-03]
dual_lm1=[2.370328e+00, 3.340228e-01, 1.277549e-01, 4.579347e-02, 1.632609e-02, 5.892160e-03]
dual_u1=[2.635130e-01, 1.324763e-02, 3.316690e-03, 8.295672e-04, 2.074367e-04, 5.186472e-05]

inv_dual_h1=inv_list(dual_h1)

slope_dual_lm1, intercept = np.polyfit(np.log(inv_dual_h1), np.log(dual_lm1), 1)
slope_dual_u1, intercept = np.polyfit(np.log(inv_dual_h1), np.log(dual_u1), 1)

# Non-conforming, 2nd order lagrange

dual_h2=[3.849491e-01, 9.623726e-02, 4.811863e-02, 2.405932e-02, 1.202966e-02, 6.014829e-03]
dual_lm2=[4.350412e+00, 6.258115e-01, 2.457169e-01, 1.156846e-01, 6.684399e-02, 5.058836e-02]
dual_u2=[3.110916e-02, 2.550737e-03, 3.877543e-04, 6.549931e-05, 1.157828e-05, 2.251468e-06]

inv_dual_h2=inv_list(dual_h2)

slope_dual_lm2, intercept = np.polyfit(np.log(inv_dual_h2), np.log(dual_lm2), 1)
slope_dual_u2, intercept = np.polyfit(np.log(inv_dual_h2), np.log(dual_u2), 1)


# non-dual
    # Finer master

    # Non-conforming, 1st order lagrange

h1=[3.849491e-01, 9.623726e-02, 4.811863e-02, 2.405932e-02, 1.202966e-02, 6.014829e-03]
lm1=[2.112048e+00, 4.176469e-02, 9.523204e-03, 2.323654e-03, 5.773288e-04, 1.441075e-04]
u1=[2.682881e-01, 1.325168e-02, 3.317933e-03, 8.298130e-04, 2.074742e-04, 5.186986e-05]

inv_h1=inv_list(h1)

slope_lm1, intercept = np.polyfit(np.log(inv_h1[2:]), np.log(lm1[2:]), 1)
slope_u1, intercept = np.polyfit(np.log(inv_h1), np.log(u1), 1)

    # Non-conforming, 2nd order lagrange

h2=[3.849491e-01, 9.623726e-02, 4.811863e-02, 2.405932e-02, 1.202966e-02, 6.014829e-03]
lm2=[1.087573e-01, 9.009581e-02, 8.275605e-02, 8.290322e-02, 8.296479e-02, 8.298276e-02]
u2=[2.322658e-02, 8.429135e-04, 1.088631e-04, 1.466607e-05, 2.616301e-06, 7.384156e-07]

inv_h2=inv_list(h2)

slope_lm2, intercept = np.polyfit(np.log(inv_h2), np.log(lm2), 1)
slope_u2, intercept = np.polyfit(np.log(inv_h2), np.log(u2), 1)

# non-dual, P2P1

h_mix=[3.849491e-01, 9.623726e-02, 4.811863e-02, 2.405932e-02, 1.202966e-02, 6.014829e-03]
lm_mix=[9.533604e-01, 7.920631e-02, 5.918822e-02, 5.799224e-02, 5.803854e-02, 5.808238e-02]
u_mix=[2.396995e-02, 8.492282e-04, 1.156252e-04, 2.025349e-05, 5.596603e-06, 1.908238e-06]

inv_h_mix=inv_list(h_mix)

slope_lm_mix, intercept = np.polyfit(np.log(inv_h_mix), np.log(lm_mix), 1)
slope_u_mix, intercept = np.polyfit(np.log(inv_h_mix), np.log(u_mix), 1)
```

```python
# plt.rc('text', usetex=True)
# plt.rc('font', family='serif')

plt_dual_lm1, =plt.plot(inv_dual_h1, dual_lm1, '-o', label='dual lm: '+ str(round(slope_dual_lm1, 3)))
plt_dual_u1, =plt.plot(inv_dual_h1, dual_u1, '->', label='dual u:'+ str(round(slope_dual_u1, 3)))

plt_lm1, =plt.plot(inv_h1, lm1, '--o', label='lm: '+ str(round(slope_lm1, 3)))
plt_u1, =plt.plot(inv_h1, u1, '-->', label='u:'+ str(round(slope_u1, 3)))

plt.yscale('log')
plt.xscale('log')
plt.xlabel('1/h', fontsize=14)
plt.ylabel('L2 Error', fontsize=14)
plt.title('Non-conforming Mesh, First Order Lagrange', fontsize=14)
plt.legend(frameon=False, fontsize=14)
plt.xlim(1e0, 1e3)
plt.ylim(1e-8, 1e2)
plt.xticks(fontsize=14)
plt.yticks([1e-8, 1e-6, 1e-4, 1e-2, 1e0, 1e2],fontsize=14)

plt.savefig('non-conform_1st_lagrange_fine_master.pdf', bbox_inches='tight')

plt.show()
```
