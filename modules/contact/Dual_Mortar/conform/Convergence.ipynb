```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def inv_list(h):
    invh=[]
    for i in h:
        invh.append(1/i)
    return invh
```

```python
# Conforming, 1st order lagrange

dual_h1=[3.535534e-01, 1.767767e-01, 8.838835e-02, 4.419417e-02, 2.209709e-02, 1.104854e-02]
dual_lm1=[1.401845e+00, 4.279261e-01, 1.123529e-01, 2.843237e-02, 7.129736e-03, 1.783789e-03]
dual_u1=[1.012553e-01, 2.552469e-02, 6.413119e-03, 1.605585e-03, 4.015452e-04, 1.003957e-04]

inv_dual_h1=inv_list(dual_h1)

slope_dual_lm1, intercept = np.polyfit(np.log(inv_dual_h1), np.log(dual_lm1), 1)
slope_dual_u1, intercept = np.polyfit(np.log(inv_dual_h1), np.log(dual_u1), 1)

# Conforming, 2nd order lagrange

dual_h2=[3.535534e-01, 1.767767e-01, 8.838835e-02, 4.419417e-02, 2.209709e-02, 1.104854e-02]
dual_lm2=[3.094306e-01, 8.204210e-02, 2.257298e-02, 6.599599e-03, 2.053744e-03, 6.721054e-04]
dual_u2=[1.180918e-02, 1.609570e-03, 2.048626e-04, 2.571729e-05, 3.218039e-06, 4.023602e-07]

inv_dual_h2=inv_list(dual_h2)

slope_dual_lm2, intercept = np.polyfit(np.log(inv_dual_h2), np.log(dual_lm2), 1)
slope_dual_u2, intercept = np.polyfit(np.log(inv_dual_h2), np.log(dual_u2), 1)

# (non-dual) Conforming, 1st order lagrange 

h1=[3.535534e-01, 1.767767e-01, 8.838835e-02, 4.419417e-02, 2.209709e-02, 1.104854e-02]
lm1=[5.872978e-01, 2.035283e-01, 7.241044e-02, 2.560984e-02, 9.074336e-03, 3.207297e-03]
u1=[1.012553e-01, 2.552469e-02, 6.413119e-03, 1.605585e-03, 4.015452e-04, 1.003957e-04]

inv_h1=inv_list(h1)

slope_lm1, intercept = np.polyfit(np.log(inv_h1), np.log(lm1), 1)
slope_u1, intercept = np.polyfit(np.log(inv_h1), np.log(u1), 1)

# (non-dual) Conforming, 2nd order lagrange 

h2=[3.535534e-01, 1.767767e-01, 8.838835e-02, 4.419417e-02, 2.209709e-02, 1.104854e-02]
lm2=[7.661794e-02, 6.801543e-02, 3.224293e-02, 1.226987e-02, 4.418867e-03, 1.569548e-03]
u2=[1.180918e-02, 1.609570e-03, 2.048626e-04, 2.571729e-05, 3.218039e-06, 4.023602e-07]

inv_h2=inv_list(h2)

slope_lm2, intercept = np.polyfit(np.log(inv_h2[3:]), np.log(lm2[3:]), 1)
slope_u2, intercept = np.polyfit(np.log(inv_h2), np.log(u2), 1)
```

```python
# plt.rc('text', usetex=True)
# plt.rc('font', family='serif')

plt_dual_lm1, =plt.plot(inv_dual_h1, dual_lm1, '-o', label='dual lm: '+ str(round(slope_dual_lm1, 3)))
plt_dual_u1, =plt.plot(inv_dual_h1, dual_u1, '->', label='dual u:'+ str(round(slope_dual_u1, 3)))

# plt_dual_lm2, =plt.plot(inv_dual_h2, dual_lm2, '-o', label='lm: '+ str(round(slope_dual_lm2, 3)))
# plt_dual_u2, =plt.plot(inv_dual_h2, dual_u2, '->', label='u:'+ str(round(slope_dual_u2, 3)))

plt_lm1, =plt.plot(inv_h1, lm1, '--o', label='lm: '+ str(round(slope_lm1, 3)))
plt_u1, =plt.plot(inv_h1, u1, '-->', label='u:'+ str(round(slope_u1, 3)))

# plt_lm2, =plt.plot(inv_h2, lm2, '--o', label='lm: '+ str(round(slope_lm2, 3)))
# plt_u2, =plt.plot(inv_h2, u2, '-->', label='u:'+ str(round(slope_u2, 3)))

plt.yscale('log')
plt.xscale('log')
plt.xlabel('1/h', fontsize=14)
plt.ylabel('L2 Error', fontsize=14)
plt.title('Conforming Mesh, First Order Lagrange', fontsize=14)
plt.legend(frameon=False, fontsize=14)
plt.xlim(1e0, 1e3)
plt.ylim(1e-8, 1e2)
plt.xticks(fontsize=14)
plt.yticks([1e-8, 1e-6, 1e-4, 1e-2, 1e0, 1e2],fontsize=14)

plt.savefig('conform_1st_lagrange.pdf', bbox_inches='tight')

plt.show()
```
