```python
import numpy as np
import pandas as pd


import matplotlib.pyplot as plt
import matplotlib.ticker as plticker

plt.rcParams['font.sans-serif'] = "Times New Roman"
# Then, "ALWAYS use sans-serif fonts"
plt.rcParams['font.family'] = "sans-serif"
```

```python
df = pd.read_csv('viscoplasticity_perfect.csv')
```

```python
df.head()
```

```python
plt.figure(0)

# plot predicted data
plt.plot(df['time']/3600, df['load']/1e6)
# plot high fidelity data
# plt.plot(np.arange(len(true_time_scaled)), true_time_scaled, 'o', markerfacecolor="None", label='Label')

# plt.xscale('log')

plt.xlabel('Time [h]',fontsize=16)
plt.ylabel('Load [MPa]',fontsize=16)
plt.xlim([-5,120])
plt.ylim([-2,30])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.legend(frameon=True, fontsize=14)

plt.savefig("perfect_visco_load_time.pdf", bbox_inches = 'tight')

```

```python
plt.figure(1)

# plot predicted data
plt.plot(df['time']/3600, df['strain'])
# plot high fidelity data
# plt.plot(np.arange(len(true_time_scaled)), true_time_scaled, 'o', markerfacecolor="None", label='Label')

# plt.xscale('log')

plt.xlabel('Time [h]',fontsize=16)
plt.ylabel(r'$\varepsilon_{yy}$ [-]',fontsize=16)
plt.xlim([-5,120])
plt.ylim([-0.05,1])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.legend(frameon=True, fontsize=14)

plt.savefig("perfect_visco_strain_time.pdf", bbox_inches = 'tight')
```
