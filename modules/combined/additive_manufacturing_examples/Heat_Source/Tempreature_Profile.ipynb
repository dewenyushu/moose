```python
import matplotlib.pyplot as plt
import numpy as np
import csv
import pandas as pd
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'Times'
```

```python
method = ['point','line']
laser_radius = '100e-3'
power = '300e-3'
directory = 'output_r'+laser_radius+'/'

# Point
fname = directory+'heat_source_out_'+power+'W_'+method[0]+'_dt25_vpp/center_temperature_line_0012.csv'
point_data=pd.read_csv(fname)

# Line
line_data = []
# for dt in [50, 75, 100]:
fname = directory+'heat_source_out_'+power+'W_'+method[1]+'_dt50_vpp/center_temperature_line_0006.csv'
line_data.append(pd.read_csv(fname))

fname = directory+'heat_source_out_'+power+'W_'+method[1]+'_dt75_vpp/center_temperature_line_0004.csv'
line_data.append(pd.read_csv(fname))

fname = directory+'heat_source_out_'+power+'W_'+method[1]+'_dt100_vpp/center_temperature_line_0003.csv'
line_data.append(pd.read_csv(fname))
```

```python
line_data[0].head()
```

```python
plt.figure(1)
    
plt.plot(point_data['x'], point_data['temp'], 'k-', linewidth=2, label ='$\hat{Q}$, $\Delta t = $ 25ms')

plt.plot(line_data[0]['x'], line_data[0]['temp'], 'k--', linewidth=2, label =r'$\bar{Q}$, $\Delta t = $ 50 ms')
plt.plot(line_data[1]['x'], line_data[1]['temp'], 'k-.', linewidth=2, label =r'$\bar{Q}$, $\Delta t = $ 75 ms')
plt.plot(line_data[2]['x'], line_data[2]['temp'], 'k:', linewidth=2, label =r'$\bar{Q}$, $\Delta t = $ 100 ms')
    

plt.xlim([0,5])
plt.ylim([0, 3500])
plt.xlabel('$x$ location [mm]',fontsize=20)
plt.ylabel('Temperature [K]',fontsize=20)
plt.xticks(np.arange(0, 6.0, 1), fontsize=16)
plt.yticks(np.arange(0, 4000, 500), fontsize=16)
plt.legend(frameon=False, fontsize=16)

# Save figure
plt.savefig('Heat_Source_Temperature_dt.pdf',bbox_inches='tight')
# plt.savefig('Stress_Strain_Correct_dG.eps',bbox_inches='tight')
# plt.savefig('Stress_Strain_Constant_E_misorientation.eps',bbox_inches='tight')
```
