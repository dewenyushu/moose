```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import csv
import pandas as pd

from matplotlib import cm

# plt.rcParams["font.family"] = "serif"

plt.rcParams['font.sans-serif'] = "Times New Roman"
# Then, "ALWAYS use sans-serif fonts"
plt.rcParams['font.family'] = "sans-serif"
```

```python
names = ['stress_xx_top','stress_xx_center','stress_xx_bot', 'stress_xx_left','stress_xx_right']
df=[]
for name in names:
    fname = 'evaluation_out_eval_'+name+'_0001.csv.0'
    df.append(pd.read_csv(fname))
```

```python
df[0].head()
```

```python
data=[]
for i in np.arange(len(df)):
    name=names[i]
    data+=[df[i]['pc_'+name].to_numpy()]
```

```python
print(data)
```

```python
fig, ax = plt.subplots(figsize=(7.2,4.45))

ax.boxplot(data)

ax.set_xlabel('Points', fontsize = '22')

ax.set_ylabel('Stress [MPa]', fontsize = '22')
ax.set_ylim([2040, 2070])

ax.tick_params(labelsize ='20')

fig.savefig('stress_dist_shoulder_loading.pdf', bbox_inches='tight' )
```
