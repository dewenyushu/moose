```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import csv
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'Serif'
```

```python
t_cyc=30;  # time for printing one layer

z_increase = 0.75;

t = np.linspace(0,89,400)  # total time = 100

phi = np.mod(t,t_cyc) # used to decide printing process on one layer

# bounded spiral curve 
a = 30
k = 1e-2
r = a*np.arctan(k*phi)


# x and y coordinates
x = r*np.cos(phi)
y = r*np.sin(phi)

z = 5.25+ (t/30).astype(int)*z_increase
```

```python
# %matplotlib qt
%matplotlib inline
# fig = plt.figure()
# ax = fig.add_subplot(111, projection='3d')
plt.plot(x,y, '-or')

# fig.savefig('printing_path_2layers.pdf', bbox_inches='tight')

# show the plot
plt.show()
```

```python
# %matplotlib qt
%matplotlib inline
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x,y,z, '-or')

# fig.savefig('printing_path_2layers.pdf', bbox_inches='tight')

# show the plot
plt.show()
```

```python
xt_data=np.vstack((t, x)).T
xt_df=pd.DataFrame(xt_data)

yt_data=np.vstack((t, y)).T
yt_df=pd.DataFrame(yt_data)

zt_data=np.vstack((t, z)).T
zt_df=pd.DataFrame(zt_data)
```

```python
xt_df.to_csv('path_t_x.csv', index=False, header=False)
yt_df.to_csv('path_t_y.csv', index=False, header=False)
zt_df.to_csv('path_t_z.csv', index=False, header=False)
```
